```python
pdbs = glob.glob("/data/lux70/plaid/artifacts/samples/5j007z42/ddim/5j007z42/*/*/structures/*.pdb")
print(len(pdb))
```

```python
import py3Dmol

for pdb in pdbs[40:50]:
    print(pdb)
    with open(pdb,"r") as f:
        view = py3Dmol.view(width=300, height=300)
        view.addModelsAsFrames(f.read())
        view.setStyle({'cartoon': {'color': {'prop': 'b', 'gradient': 'roygb', 'min': 50, 'max': 90}}})
        view.addSurface(py3Dmol.VDW, {'opacity': 0.7, 'colorscheme': {'prop': 'b', 'gradient': 'roygb', 'min': 50, 'max': 90}})
        view.zoomTo()
        view.show()
```
